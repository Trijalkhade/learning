**<h1>Using Built-in Functions</h1>**

```python
import numpy as np
```

```python
array = np.array([])
```

```python
n = int(input("Enter the number of elements: "))
```

```python
for i in range(n):
  num = float(input(f"Enter number {i+1}: "))
  array = np.append(array,num)
```

```python
# Calcculating the minimum maximum and mean value
array_min = np.min(array)
array_max = np.max(array)
array_mean = np.mean(array)
```

```python
# printing the minimum maximum and mean value
print(f"Minimum value: {array_min}")
print(f"Maximum value: {array_max}")
print(f"Mean value: {array_mean}")
```

**<h2>Defining function</h2>**

```python
def min(array):
  max_val = array[0]
  n = len(array)
  for i in range(n - 1):
    if array[i] > array[i+1]:
      min_val = array[i+1]
  return min_val
```

```python
def max(array):
  max_val = array[0]
  n = len(array)
  for i in range(n - 1):
    if array[i] < array[i+1]:
      max_val = array[i+1]
  return max_val
```

```python
def mean(array):
  mean_val = array[0]
  n = len(array)
  for i in range(n - 1):
    mean_val += array[i+1]
  mean_val /= n
  return mean_val
```

```python
# printing the minimum maximum and mean value
print(f"Minimum value: ", min(array))
print(f"Maximum value: ", max(array))
print(f"Mean value: ", mean(array))
```
