# Recursive Fibonacci

```python
def fibonacci(n):
  if n < 0:
    print("Invalid input. Please enter a positive integer.")
    return 0
  elif n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
n = int(input("Enter the value of n: "))

while n>0:
  print(fibonacci(n))
  n-=1
```

# Fibonacci logic

```python
i,a,b,show = 0,0,1,0

n = int(input("Enter the value of n: "))

print("0")
print("1")
while i<n:
   show=a+b
   a=b
   b=show
   print(show)
   i+=1
```
