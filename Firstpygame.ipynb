```python
import random
random.random()
count = 0
x = random.randrange(1, 10)
y=11
while x!=y and count<3:
  y = int(input("Guess a number between 1 -10:"))
  if x==y:
    print("Correct")
    print("Number of chances:", count)
  elif x<y:
    print("Wrong")
    print("Hint: Try a smaller number")
    count+=1
  elif x>y:
    print("Wrong")
    print("Hint: Try a bigger number")
    count+=1
if count>2:
  print("You lost the game")
```

```python
x = "Hello World"
print("string:", x)
x = 20
print("integer:", x)
x = 20.5
print("float:", x)
x = 1j
print("complex:", x)
x = ["apple", "banana", "cherry"]
print("list:", x)
x = ("apple", "banana", "cherry")
print("tuple:", x)
x = range(6)
print("range:", x)
x = {"name" : "John", "age" : 36}
print("dictionary:", x)
x = {"apple", "banana", "cherry"}
print("set:", x)
x = True
print("boolean:", x)
x = b"Hello"
print("byte:", x)
x = bytearray(5)
print("bytearray:", x)
x = memoryview(bytes(5))
print("memoryview:", x)
x = None
print("none:", x)
```
