```python
n = int(input("Enter the number of elements:"))
numbers = []
for i in range(n):
  num = int(input("Enter a number: "))
  numbers.append(num)
```

```python
even = []
odd = []
for num in numbers:
  if num%2 == 0:
    even.append(num)
  else:
    odd.append(num)
```

```python
even_odd_dict = {'EVEN': even, 'ODD': odd}
print(even_odd_dict)
```
