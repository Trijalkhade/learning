```python
# Check to find palindrome
string = input()
if string == string[::-1]:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
# Removing of consecutive duplicate
str = input()
str1 = ""
for i in range(len(str)-1):
  if (str[i] != str[i+1]):
    str1+=str[i]
  else:
    str1+=str[i]

print(str1)
```

```python
# Removing of all duplicate
str = input()
for i in range(len(str)-1):
  for j in range(i+1,len(str)):
    if str[i] == str[j]:
      str = str[j] + str.replace(str[j],"")
      break
print(str)
```
