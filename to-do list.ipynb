```python

#REFERENCE
import json
import datetime

class Task:
    def __init__(self, description, priority="Normal", due_date=None):
        self.description = description
        self.priority = priority
        self.due_date = due_date
        self.completed = False

    def __str__(self):
        status = "[x]" if self.completed else "[ ]"
        due_date_str = f", Due: {self.due_date.strftime('%Y-%m-%d')}" if self.due_date else ""
        return f"{status} {self.description} (Priority: {self.priority}{due_date_str})"

def add_task(tasks):
    description = input("Enter task description: ")
    priority = input("Enter priority (High, Normal, Low): ").capitalize()
    due_date_str = input("Enter due date (YYYY-MM-DD, optional): ")
    due_date = datetime.datetime.strptime(due_date_str, "%Y-%m-%d").date() if due_date_str else None
    tasks.append(Task(description, priority, due_date))
    print("Task added!")

def remove_task(tasks):
    view_tasks(tasks)
    if not tasks:
        return

    try:
        task_number = int(input("Enter the number of the task to remove: ")) - 1
        if 0 <= task_number < len(tasks):
            del tasks[task_number]
            print("Task removed!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input.")

def view_tasks(tasks):
    if not tasks:
        print("Your to-do list is empty.")
        return

    for i, task in enumerate(tasks):
        print(f"{i + 1}. {task}")

def mark_complete(tasks):
    view_tasks(tasks)
    if not tasks:
        return

    try:
        task_number = int(input("Enter the number of the task to mark complete: ")) - 1
        if 0 <= task_number < len(tasks):
            tasks[task_number].completed = True
            print("Task marked complete!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input.")

def search_tasks(tasks):
    search_term = input("Enter search term: ")
    matching_tasks = [task for task in tasks if search_term.lower() in task.description.lower()]
    view_tasks(matching_tasks)

def save_tasks(tasks, filename="todo.json"):
    task_data = [{
        "description": task.description,
        "priority": task.priority,
        "due_date": task.due_date.isoformat() if task.due_date else None,
        "completed": task.completed
    } for task in tasks]
    with open(filename, "w") as f:
        json.dump(task_data, f)
    print("Tasks saved!")

def load_tasks(filename="todo.json"):
    try:
        with open(filename, "r") as f:
            task_data = json.load(f)
        tasks = [Task(
            data["description"],
            data["priority"],
            datetime.date.fromisoformat(data["due_date"]) if data["due_date"] else None
        ) for data in task_data]
        for task, data in zip(tasks, task_data):
            task.completed = data["completed"]
        return tasks
    except FileNotFoundError:
        return []

def main():
    tasks = load_tasks()

    while True:
        print("\nChoose an action:")
        print("1. Add task")
        print("2. Remove task")
        print("3. View tasks")
        print("4. Mark task complete")
        print("5. Search tasks")
        print("6. Save tasks")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_task(tasks)
        elif choice == "2":
            remove_task(tasks)
        elif choice == "3":
            view_tasks(tasks)
        elif choice == "4":
            mark_complete(tasks)
        elif choice == "5":
            search_tasks(tasks)
        elif choice == "6":
            save_tasks(tasks)
        elif choice == "7":
            print("Goodbye!")
            break
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main()
```
